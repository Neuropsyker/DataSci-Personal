```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',178)
```

```python
# Read in notebook and convert the index to a datetime 
df = pd.read_csv("../output/final_base.csv", index_col='Date')
df.head()
df.index = pd.to_datetime(df.index)
df.index.dtype
```

```python
# Find all the missing values for each column names 
df.isna().sum().sort_values(ascending=False)
```

```python
# Impute or delete or decide what to do with the missing values 
# find the minimum date

# missing values per row
missing_values = df.isna().sum(axis=1)

# Finding the dates where there are no missing values 
df.index[missing_values<=0]

# Finding the dates where there are no missing values 
# 10 missing values in a row = 27/07/2020
# 18 missing values in a row = 15/04/2020
# for loop to find the threshold of the number of missing values 
for i in range(5,31, 1):
    print(i, df.index[missing_values<=i].min())
```

```python
# start date we've decided = 2020-01-01 00:00:00

df_20_to_24 = df[df.index>='2020-01-01']
# df_20_to_24.describe().T.sort_values('count',ascending=True) # This is why we dropped these columns

df_20_to_24 = df_20_to_24.drop(['EURO_Index_Vol.', 'EURO_Index_Low', 'EURO_Index_High', 'EURO_Index_Open', 'USD_Index_Vol.', 'USD_Index_Open', 'USD_Index_High', 'USD_Index_Close_Price','EURO_Index_Close_Price','USD_Index_Low'], axis=1)
df_20_to_24.describe().T.sort_values('count',ascending=True)
```

```python
df_20_to_24.head()

# forward filling the values for the global market 
column_names_volume = []

for i in df_20_to_24.columns:
        if 'Vol.' in i:
            column_names_volume.append(i)
        
```

```python
df_20_to_24[column_names_volume] = df_20_to_24[column_names_volume].fillna(0, axis=1)
df_20_to_24[column_names_volume].info()

# price but with ffill()
df_20_to_24 = df_20_to_24.ffill()
df_20_to_24 = df_20_to_24.bfill()
```

```python
# Identify if we still have missing values: 
missing_values = df_20_to_24.describe().T.sort_values('count',ascending=True)

df_20_to_24.to_csv('../output/df_2020_2024.csv')
```
