## Mean

- The mean of a population is defined as mu (greek letter)
- The mean of a sample is called x-bar
- You get the mean by adding up all the components and then dividing by the number of components
- The slight issue of mean is that it is easily impacted by outliers

```python
import pandas as pd

df = pd.DataFrame()
df['weatherinNYC'] = [50, 43, 47, 50, 155, 40] 
df['weatherinChicago'] = [22, 12, 12, 34, 23, 34]



print(f"The average weather in Chicago is {df['weatherinChicago'].mean().round()} and the average weather in NYC is {df['weatherinNYC'].mean().round()}.")
```

The outlier in the NYC, skews the mean to be significantly higher. 

In order to manage this we look at the median, the second of the averages - the median. In order to calculate the median we do the following:

- Order the rows from lowest values to highest values
- Find the middle value; this is the median value

```python
print(f"The median weather in Chicago is: {df['weatherinChicago'].median()}, the median weather in NYC is: {df['weatherinNYC'].median()}")

# Hence this shows that the median is unaffected by extreme values (outliers)
```

## Mode

The mode is a way of identifying the most common value in a dataset.

## Skewness

Skewness helps explain whether the data is concentrated on one side or on another.

The mean > median - therefore there is a positive skew and so this means that the outliers lie on the RHS. 

- A dataset with the mean = mode = median will be a balanced dataset. Therefore, you should get a symmetrical distribution.
- A dataset with the mean < median and the mode will have a negative skew (i.e. a left skew - this is because the outliers will be to the left)

- The skewness tells us where most of the data sits
