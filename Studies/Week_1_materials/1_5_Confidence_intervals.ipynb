### Normal distribution and what this means

If we say that a variable is normally distributed we are making the assumption that the majority of the observations are around the mean.

There are the confidence intervals, for which there is the lower limit and the upper limit.

A 95% confidence interval would imply that there is 95% confident that the true population mean falls within this interval.



### Confidence and the relationship to confidence intervals

- When our confidence is lower; the confidence interval is smaller. I.e. to have a confidence of 90%, the confidence interval will be smaller
- If we have a higher confidence; let's say 99% then our confidence interval will be larger as we have to be 99% confident that the values fall within this range.


Let's say you were using confidence levels to highlight the confidence of people being between different ages:

- The chance that people are between 20 and 55 - confidence = 85%
- The chance that people are between 18 and 70 - confidence = 95%
- The chance that people are between 0 - 129 - confidence = 100%

The 100% confidence interval is not very useful as it has to include all the different values.

### The null hypothesis vs the alternative hypothesis

- The Null hypothesis is denoted as H0
- The alternative hypothesis is everything else


Let's say you were interested in seeing the average salary of computer scientists. And we thought that the average price was $130,000. 

The alternative hypothesis would be that the average salary was not equal to $130,000.

Whilst the null hypothesis would be that the average salary was $130,000.

The null hypothesis is accepted until it is deemed that it is incorrect. The way to test this is to design a test. The null hypothesis is the thing that we are trying to reject.
