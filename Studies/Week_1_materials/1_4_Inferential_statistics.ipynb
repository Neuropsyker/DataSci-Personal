##### Shifting the mean

#### Shifting the standard deviation, whilst controlling for the mean:

- By increasing the standard deviation (the spread around the mean), whilst keeping the mean the same results in a graph with a flatter curve and more data in the tails and less data in the center
- By decreasing the standard deviation and controlling for the mean, results in a taller curve, with more data in the middle and thinner tails



#### Standardisation

- If a data variable has a mean of mu and a standard deviation of sigma**2; the process of standardisation is making the mean = 0 and the standard deviation = 1.
- A normal distribution can also be normalised, the result is a standard normal distribution
- In order to get the standard normal distribition:

x - mean(mu) / standard deviation

### The standard normal distribution



```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

series_a = [1, 2, 2, 3, 3, 3, 4, 4, 5]
mean = np.mean(series_a)
std_dev = np.std(series_a)

print(f"The mean for series_a = {mean}. The standard deviation for series_a = {round(std_dev,2)}.")


series_a_minus_mean = series_a - mean
series_a_minus_mean_standard = series_a_minus_mean  / np.std(series_a_minus_mean)
mean_series_a = np.mean(series_a_minus_mean)
std_dev_a = np.std(series_a_minus_mean)

print(f"The mean for series_a = {mean_series_a}. The standard deviation for series_a = {round(np.std(series_a_minus_mean_standard),2)}.")


series_a_minus_normalised = (series_a - mean)/std_dev_a
mean_series_a_minus_normalised = np.mean(series_a_minus_normalised)
std_dev_a_normalised = np.std(series_a_minus_normalised)

# print(f"The mean for series_a = {mean_series_a_minus_normalised}. The standard deviation for series_a = {round(std_dev_a_normalised,2)}.")


sns.histplot(data = series_a, kde=True, label='series_a')
sns.histplot(data = series_a_minus_mean, kde=True, label='series_a_minus_mean')
sns.histplot(data = series_a_minus_normalised, kde=True, label='series_a_normalised', multiple="stack")
plt.legend(loc='best')
```

### Central Limit Therom

- No matter the underlying distribution, the sampling mean distribution will approximate to a normal distribution. 
- The central limit therom allows us to solve problems, make inferences and use the normal distribution even when the population is not normally distributed


You have a population with any shape of distribution (it could be normal, uniform, skewed, or something else). If you repeatedly draw random samples of a sufficiently large size from that population and calculate the mean of each sample, the distribution of those sample means will tend to follow a bell-shaped, normal distribution, as the sample size increases.

```python
import numpy as np 

scores = [2747, 1275, 1423, 1172, 1671,  223, 1722, 2244, 2946, 2350,  584,
       1952, 2494, 1111,  774,  132,  610, 1803, 1786, 2975,  526, 1232,
       1831, 1043, 2559, 2252, 2808, 1147, 2919, 1946,  178, 2752, 1373,
        491, 1171,  525, 1070,  871,   91, 1618,  998, 1732,  935,  965,
       1622,  304, 2963,   56, 2285, 1650,  249,  821, 2343,   62, 1700,
       1493,  418, 2609, 2966, 1437, 2588, 2145, 1853, 1625, 2418, 1164,
       1277, 2457,  412, 1350, 1815,  449, 1252, 2719, 1044, 2170, 2879,
       1815,  975,  673, 2418, 1566, 2756, 1499, 1138, 2840, 1213,  358,
        820, 1570, 1183,  547,  215, 1062, 1144,  724, 1864, 1397,  680,
       2473]

sns.histplot(scores, kde=True, label='Original data distribution')

np.random.seed(40)
random_indexes_sets = [np.random.randint(0,100,5) for _ in range(100)]
random_indexes_sets

mean = []

# Iterate over each set of random indexes in random_indexes_sets
for indexes_set in random_indexes_sets:
    total = 0
    
    # Iterate over the indexes in the current set
    for i,j in enumerate(indexes_set):
        total += scores[j]
    
    # Calculate the mean for the current set and append it to the 'mean' list
    set_mean = total / len(indexes_set)
    mean.append(set_mean)


sns.histplot(mean, kde=True, label='Sample mean distributin')
plt.title('The Central Limit Therom')
plt.legend(loc='best')
```

### Standard Error:

- The standard error is calculated as the standard deviation of the sample / square_root(n)
- The standard error decreases when the sample size increases
- The standard error shows variability of the means of the samples 
- It shows how well you have approximated the true mean
