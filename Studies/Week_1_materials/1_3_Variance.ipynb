## What is variance?

- Variance measures the dispersion of a set of data points around the mean
- There are 2 different formulas for variance - there is the population variance and the sample variance

### Population variance vs Sample variance

- The biggest difference between the 2 formulas is the denominator
- In the top formula (Population variance), we divide by N 
- In the bottom formula (Sample variance), we divide by N-1

- The closer the values are to the mean that smaller the numerator
- By squaring the difference we magnify the difference between the value and the mean
- By squaring the values we remove the potential for negative values; which makes sense as variance is looking at distance from the mean

```python
## Population variance

population = [3,5,6,7,19]
mean = sum(population) / len(population)
mean

# variance = (3-mean)**2 + (5-mean)**2 ... this will take time lets write this out pythonically 

val = 0
for i in population:
    val += (i-mean)**2

print(f"Variance of the population = {val/ len(population)}")

# Confirm value is correct by using the inbuild library of numpy
import numpy as np
np.var(population)
```

```python
# Sample Variance 

print(f"Variance of the population = {val/ (len(population)-1)}")
```

The sample variance has corrected upwards in order to reflect the higher potential variability

## Standard deviation

- The population standard deviation is seen as the square root of the population Variance
- The sample standard deviation is seen as the square root of the sample standard Variance

- The standard deviation shows how dispersed the data is in relation to the mean
- Low / small standard deviations show that the data is clustered/ centered around the mean 
- High / large standard deviations show that the data is more spread out

### Coefficient of variation

- Comparing the standard deviations of 2 datasets is pointless
- Comparing the coefficients of variations is easier to compare

- The formula for the coefficient of variation is:

The Coefficient of Variation = Standard Deviation / mean 

### Covariance

- There is a sample and a population formula
- The sample formula for the covariance is: 

(x-x_bar) * (y-y_bar) / n-1


- Covariance tells us about the direction that the variables move in:
1. If the covariance > 0 then the variables move together 
2. If the covariance < 0 then the 2 variables move in opposite direction 
3. If the covariance = 0 then the 2 variables are independent

- Covariance can be of a huge scale and so it can be hard to interpret the results and so it can be hard to interpret the results

### Correlation 

- Perfect positive correlation: This means that the entire variability of one variable is explained by the other
- Correlation of 0: means that the 2 variables are absolutely independent from one another
- Perfect negative correlation: This means that as one variable increases the other decreases
