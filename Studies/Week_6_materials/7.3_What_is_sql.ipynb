### How SQL works 

- SQL is a method of querying databases
- The first querying key word is "SELECT" 
- You then specify the database.table that you are interested in by using the keyword "FROM"

- Similar to python and excel you can specify specific scenarios 
- Using the keyword "WHERE"


#### SQL query format (when querying)

- "SELECT * FROM <TABLE_NAME>"
- "SELECT * FROM <TABLE_NAME> WHERE <CONDITION>" 

### SQL joins and merges 

1. There are multiple different types of joins
- Left joins
- Right joins
- Inner joins 
- Outer joins

2. These joins allow you to take one data table and join it to another data table 

<!-- Left joins -->
<!-- This takes all the rows in table A and then joins them to rows in column be where the keys in table A and B match -->
SELECT <COLUMNS FROM TABLE A: A.COLUMN NAME>, <COLUMNS FROM TABLE B: B.COLUMN NAME>
FROM TABLE_A A
LEFT JOIN TABLE_B B 
ON A.KEY = B.KEY 

<!-- RIGHT joins -->
<!-- This takes all the rows in table B and then right joins them to rows in table A where the keys in table A and B match -->
SELECT <COLUMNS FROM TABLE A: A.COLUMN NAME>, <COLUMNS FROM TABLE B: B.COLUMN NAME>
FROM TABLE_A A
RIGHT JOIN TABLE_B B 
ON A.KEY = B.KEY
