## MSE and $R^2$

The sum of squared residuals looks at the:

$ Residual = (Observed - Predicted)^2 $

This therefore, means that it is a function of the the number of values being observed and predicted (i.e the amount of data that you have).
 

The mean squared error corrects for this with the following equation:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$


$$
\text{MSE} = \frac{\sum_{i=1}^{n} SSR}{\text{Number of observations}}
$$


The problem with MSE is that it can be difficult to interpret because it depends on the scale of the data. Instead we use the $R^2$ value. 

### $R^2$

- How $R^2$ works is that it compares the MSE around the mean y-axis value and then it calculates the MSE around the average height. 
- We then compare the MSE around the model 
- $R^2$ then give us a percentage of how much the predictons improved by using the model we're interested in instead of the mean. 
- $R^2$ ranges from 0 to 1. 
- The closer the value is to 1, the better the model fits the data relative to the mean y-axis value.

### How to calculate $R^2$
1. Calculate the SSR / MSE for the mean 
2. Calculare the SSR for / MSE the fitted line 
3. Then plug into this formula:

$$
R^2 = \frac{\text{SSR(mean) - SSR(fitted line)}}{\text{SSR (mean)}}
$$





- This formula works is that it tells us how much the SSRs reduced once we fitted the line
- N.b. The most common $R^2$ formula is to compare to the mean, however, it doesn't always have to be a mean! It can be used to compare it to a different function. 
- When we are using the mean as the function we can't ever get a negative value for the $R^2$ value 
- $R^2$ is based on the sum of square squared residuals around the mean and the fitted line
- If you can caluclare the Pearson's correlation coefficient for a relationship between 2 things then the square of that coefficient is equal to $R^2$

### P-values

- P-values allow us to quantify how much confidence we should have in the $R^2$ value. 
- P-values give us a measure of confidence in the results from a statistical analysis 
- P-values are numbers between 0 and 1 
- We need to determine the null hypothesis and then we need to decide whether to reject or accept the null hypothesis
