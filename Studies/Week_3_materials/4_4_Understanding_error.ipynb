Linear regression even the simple kind requires us to quantify how well the model is performing. Let's say we are building a model that classifies height from weight. As we know from a set of data there are multiple different "lines" that can be fit to the data. 


- For us to understand which "line fits" best we use SSR. 


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
height = np.array([150, 155, 160, 165, 170, 175, 180, 185])
weight = np.array([50, 49, 60, 68, 70, 75, 82, 78])

# First linear regression
m1, b1 = np.polyfit(height, weight, 1)

# Second linear regression (using a different method or modifying parameters)
# For demonstration, I'll use the same method but with a slight modification
m2, b2 = np.polyfit(height, weight * 1.05, 1)  # Slightly modified weights

# Create a range of values for height to plot the regression lines
height_range = np.linspace(height.min(), height.max(), 100)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(height, weight, color='blue', label='Data points')
plt.plot(height_range, m1 * height_range + b1, color='red', label='Linear Regression 1')
plt.plot(height_range, m2 * height_range + b2, color='green', label='Linear Regression 2')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight with Two Lines of Best Fit')
plt.legend()
plt.show()
```

#### How do we know which line is better?

- A way of quantifying which model is better is by using the sum of squared residuals
- The residual is the observed - predicted
- The smaller the residuals the better the model fits


$$
\text{SSR} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

- y_i is the observed 
- y_hat is the predicted
