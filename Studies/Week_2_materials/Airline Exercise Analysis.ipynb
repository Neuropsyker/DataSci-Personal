```python
# Grouping flight length distances as short, medium, and long-haul.
df = pd.read_csv("airline.csv")
flight_distance_bins=[31, 999, 3499, 5000] 

# Using cut function to split flight distance scores into three categories:

flight_distance_category = pd.cut(df['Flight Distance'], flight_distance_bins,
       labels=['Short Haul', 'Medium Haul', 'Long Haul'])

# Printing this data in a concise, readable format

flight_distance_counts = flight_distance_category.value_counts().reset_index()
ax=sns.countplot(data=df, x=flight_distance_category)
ax.set_title("Customers Grouped by Flight Distance")
plt.show()
```

```python
flight_distance_counts = flight_distance_category.value_counts(normalize=True).reset_index()
print(flight_distance_counts)
```

Now let's look at how in-flight variables correlate with overall satisfaction.

```python
df = pd.read_csv("airline.csv")
correlation_matrix = df[['Overall Satisfaction', 'In-flight Wifi Service', 'Food and Drink', 'On-board Service', 'Seat Comfort', 'Leg Room Service', 'Cleanliness', 'In-flight Entertainment']].corr()
sns.heatmap(correlation_matrix, annot=True, annot_kws={'fontsize': 8})
plt.show()
```

What can we tell from the above? It's probably a bit of a Captain Obvious statement but all of the in-flight satisfaction variables have a positive correlation with mean in-flight satisfaction scores. It appears that entertainment has the highest correlation and therefore perhaps the strongest impact on passenger satisfaction. Seat comfort also has an especially strong positive correlation with satisfaction. Wi-Fi and leg room had the lowest correlation coefficient of the bunch, but not weak! Now how does it compare to in-flight satisfaction vs. airport variable satisfaction?

```python
# Looking at airport variables and how they correlate to average satisfaction and airport service satisfaction
df = pd.read_csv("airline.csv")
correlation_matrix = df[['Overall Satisfaction', 'Airport Satisfaction', 'Mean In-flight Satis.']].corr()
sns.heatmap(correlation_matrix, annot=True, annot_kws={'fontsize': 8})
plt.show()
```

From the above, we can see that In-flight Satisfaction has a higher correlation than Airport satisfaction with Overall Satisfaction, but both correlations are strong positive. Of the airport service satisfaction variables, baggage handling had the highest correlation with overall satisfaction, followed by check-in service and gate location. 

This is good because it allows us to reject a previous hypothesis that airport variables outside the airline's control would contribute more to passenger satisfaction than in-flight variables that would be within the airline's purview.
